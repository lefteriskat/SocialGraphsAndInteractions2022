```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
#source https://www.geeksforgeeks.org/scrape-imdb-movie-rating-and-details-using-python/
#if you want to play you will need 
#https://codebeautify.org/htmlviewer   make html better
```

```python
url = 'https://www.imdb.com/search/title/?series=tt2861424&sort=user_rating,desc&count=250&view=advanced'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
episodes = soup.find_all('div',class_='lister-item mode-advanced')
title = episodes[0].find('small',class_='text-primary unbold').find_next().string
rating = episodes[0].find('div',class_='inline-block ratings-imdb-rating').strong.string
number = episodes[0].find('span',class_='lister-item-index unbold text-primary').string
```

```python
rating
```

```python
# create a empty list for storing
# movie information
list = []
 
# Iterating over movies to extract
# each movie's details
for index in range(0, len(episodes)):
     
    episode_name = episodes[index].find('small',class_='text-primary unbold').find_next().string
    
    if (index == 57 or index == 58 or index == 59 or index == 60):
        rating = 0
    else:
        rating = episodes[index].find('div',class_='inline-block ratings-imdb-rating').strong.string
        
    data = {"movie_title": episode_name,
            "rating": rating
            }
    list.append(data)
```

```python

#saving the list as dataframe
#then converting into .csv file
df = pd.DataFrame(list)
```
